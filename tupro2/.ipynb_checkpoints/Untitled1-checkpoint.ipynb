```python
import numpy as np
import matplotlib.pyplot as plt
import math
import random
import copy
```

```python
# Importing Dataset
Dataset = np.genfromtxt('Tugas 2 ML Genap 2018-2019 Dataset Tanpa Label.csv', delimiter=",")
Dataset = np.asarray(Dataset)
```

```python
def normalize(Dataset):
    Dataset[:,0] = (Dataset[:,0]- min(Dataset[:,0])) / (max(Dataset[:,0]) - min(Dataset[:,0]))
    Dataset[:,1] = (Dataset[:,1]- min(Dataset[:,1])) / (max(Dataset[:,1]) - min(Dataset[:,1]))
    return Dataset
#Dataset = normalize(Dataset)
neuron = np.random.randint(0, 20, size=(100, 2)) #/ 255
plt.plot(neuron[:,0],neuron[:,1])
plt.plot(Dataset[:,0], Dataset[:,1],'ro')
plt.show()
#Dataset = np.reshape(Dataset, (Dataset.shape[0],1,Dataset.shape[1]))
```

```python
def getEuclidean(p, q):
        return np.sqrt(np.sum((q-p)**2))

def train(dataset):
    learning_rate = 0.1
    t_learning_rate = 2

    radius = 2
    t_radius = 2

    iteration = 10
    convergen_criterion = 1 * 10**-15
    
    #for t in range(iteration):
    for t in range(iteration):
        print("=========")
        x = random.choice(dataset)
        
        win_neuron = random.choice(neuron)
        #print(win_neuron)
        for i in neuron:
            if getEuclidean(x, i) < getEuclidean(x, win_neuron): 
                win_neuron = i
        
        neighbors = []
        neighbors.append(win_neuron)
        for j in neuron:
            if(getEuclidean(win_neuron,j) < radius):
                neighbors.append(j)
        
        for j in neighbors:   
            s = getEuclidean(win_neuron,j)
            phi = np.exp(-(s*s) / (2*radius*radius))
            dWeight = learning_rate * phi * (x-j)
            neuron[np.nonzero(neuron == j)[0][0]] += dWeight
           
        if (radius > convergen_criterion):
            print("iterasi: ",t)
            
          
        learning_rate *= np.exp(-t/ 2)
        radius *= np.exp(-t/2)
            
print(neuron[0:20])

train(Dataset)
print(neuron[0:20])
plt.plot(neuron[:,0],neuron[:,1])
plt.plot(Dataset[:,0], Dataset[:,1],'ro')
plt.show()
```
